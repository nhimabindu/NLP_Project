```python
opportunity_keywords = [
    "growth",
    "investment",
    "expansion",
    "renewable",
    "partnership",
    "innovation",
    "ai",
    "automation",
    "new market"
]

risk_keywords = [
    "tariff",
    "risk",
    "threat",
    "delay",
    "shortage",
    "regulation",
    "competition",
    "disruption"
]

trend_keywords = [
    "ai",
    "automation",
    "electric",
    "battery",
    "sustainability",
    "renewable",
    "digital"
]

def detect_opportunity(text):
    text = str(text).lower()
    return sum(keyword in text for keyword in opportunity_keywords)

def detect_risk(text):
    text = str(text).lower()
    return sum(keyword in text for keyword in risk_keywords)

def detect_trend(text):
    text = str(text).lower()
    return sum(keyword in text for keyword in trend_keywords)
```

```python
combined_df["opportunity_score"] = (
    combined_df["clean_text"].apply(detect_opportunity)
)

combined_df["risk_score"] = (
    combined_df["clean_text"].apply(detect_risk)
)

combined_df["trend_score"] = (
    combined_df["clean_text"].apply(detect_trend)
)

print("Done")
```

```python
#Trend Classification
def classify_trend(text):

    text = str(text).lower()

    if "robot" in text or "automation" in text:
        return "Automation"

    elif "battery" in text:
        return "Battery Logistics"

    elif "renewable" in text or "energy" in text:
        return "Renewable Energy"

    elif "sustainability" in text or "gogreen" in text:
        return "Sustainability"

    elif "ai" in text:
        return "Artificial Intelligence"

    return "Other"

dhl_df["trend_category"] = (
    dhl_df["clean_text"]
    .apply(classify_trend)
)
```

```python
#Then Build Opportunity Categories
def classify_opportunity(text):

    text = str(text).lower()

    if "energy" in text or "renewable" in text:
        return "Energy Logistics"

    elif "battery" in text:
        return "Battery Supply Chain"

    elif "automation" in text or "robot" in text:
        return "Warehouse Automation"

    elif "ai" in text:
        return "AI Logistics"

    elif "partnership" in text:
        return "Strategic Partnership"

    return "Other"

dhl_df["opportunity_category"] = (
    dhl_df["clean_text"]
    .apply(classify_opportunity)
)
```

```python
#Then Build Risk Categories
def classify_risk(text):

    text = str(text).lower()

    if "tariff" in text:
        return "Trade Tariff"

    elif "regulation" in text:
        return "Regulatory Risk"

    elif "competition" in text:
        return "Competitive Risk"

    elif "delay" in text or "disruption" in text:
        return "Supply Chain Disruption"

    return "Other"

dhl_df["risk_category"] = (
    dhl_df["clean_text"]
    .apply(classify_risk)
)
```
