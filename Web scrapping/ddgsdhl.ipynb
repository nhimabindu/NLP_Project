```python
from ddgs import DDGS
import pandas as pd

results = []

queries = [
    "DHL logistics news",
    "DHL supply chain",
    "DHL automation",
    "DHL sustainability"
]

with DDGS() as ddgs:
    for query in queries:
        for r in ddgs.text(query, max_results=50):
            results.append(r)

ddgs_df = pd.DataFrame(results)

print("Before duplicates:", len(ddgs_df))

ddgs_df = ddgs_df.drop_duplicates(subset=["href"])

print("After duplicates:", len(ddgs_df))

ddgs_df.to_csv("ddgs_dhl.csv", index=False)

print("Saved successfully!")
```
