```python
# Save Articles Automatically
from newsapi import NewsApiClient
import json

# Your API key
newsapi = NewsApiClient(api_key="5d279bce96434d7483abf8c7cb594d0d")

# Search DHL-related news
articles = newsapi.get_everything(
    q="DHL",
    language="en",
    sort_by="publishedAt",
    page_size=100
)

news_data = []

for article in articles["articles"]:
    news_data.append({
        "title": article["title"],
        "source": article["source"]["name"],
        "published_at": article["publishedAt"],
        "url": article["url"],
        "description": article["description"]
    })

# Save to JSON
with open("dhl_news.json", "w", encoding="utf-8") as f:
    json.dump(news_data, f, indent=4)

print(f"Collected {len(news_data)} articles.")
```

```python
# Verify Your Data

import json

with open("dhl_news.json", "r", encoding="utf-8") as f:
    data = json.load(f)

print("Articles:", len(data))

sources = set()

for article in data:
    if article["source"]:
        sources.add(article["source"])

print("Unique Sources:", len(sources))
print(list(sources)[:20])
```
